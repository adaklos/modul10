```python
import random
from random import randint
from datetime import date

class Movie:
    # Przechowuje filmy i seriale w jednej liście
    library_list = []
    def __init__(self, title, year, genre, views):
        self.title = title
        self.year = year
        self.genre = genre
        self.views = views
        self.library_list.append(self)         
        
        self.current_views = views       
        
    # Napisz funkcje get_movies oraz get_series, które będą filtrować listę i zwracać odpowiednio
    #tylko filmy oraz tylko seriale          
    def get_movies():
        movies_list = []
        for x in Movie.library_list:
            if x.__class__ == Movie:
                movies_list.append(x)                
        return movies_list
    
    def get_series():
        series_list = []
        for x in Movie.library_list:
            if x.__class__ == Series:
                series_list.append(x)
        return series_list

    # Filmy i seriale mają metodę play, która zwiększa liczbę odtworzeń danego tytułu o 1
    def play(self,step=1):
        self.current_views += step
    
    # Po wyświetleniu filmu jako string widoczne są tytuł i rok wydania np. “Pulp Fiction (1994)”
    def __str__(self):
        return f'{self.title} ({self.year})'
    
    def __repr__(self):
        return f"Movie (title={self.title} year={self.year}, genre={self.genre}, views={self.views})"
    
    # Napisz funkcję search, która wyszukuje film lub serial po jego tytule.
    @property
    def titlea(self):
        return f'{self.title},{self.year}'

    @staticmethod
    def search():
        print()
        user_input = input("What title are you searching for? ")
        print()

        result = [i for i in Movie.library_list
            if user_input in (i.title)]

        print(f"Matching titles: {', '.join(i.titlea for i in result)}")
    
    
    # Napisz funkcję, która uruchomi generate_views 10 razy
    def ten_times(func):
        def wrapper(*args, **kwargs):
            for i in range(10):
                func(*args, **kwargs)
        return wrapper
              
              
    #Napisz funkcję generate_views, która losowo wybiera element z biblioteki,
    # a następnie dodaje mu losową (z zakresu od 1 do 100) ilość odtworzeń.
    @ten_times    
    def generate_views():
        random_title = random.choice(Movie.library_list)
        step = randint(0,100)
        ind = Movie.library_list.index(random_title)
        Movie.library_list[ind].views += step
        
              
    # Napisz funkcję top_titles(), która zwróci wybraną ilość najpopularniejszych tytułów z biblioteki          
    def top_titles(x):
        top_views = sorted(Movie.library_list, key=lambda movie: movie.views, reverse = True)
        return(top_views[:x])
   
    
        
class Series(Movie):
    def __init__(self, episode, season, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.episode = episode
        self.season = season
              
    def get_series():
        series_list = []
        for x in Movie.library_list:
            if x.__class__ == Series:
                series_list.append(x)
        return series_list
    
    # Po wyświetleniu serialu jako string pokazują się informacjeo konkretnym odcinku,
    def __str__(self):
        return f'{self.title} SO{self.season}E0{self.episode}'
    
    def __repr__(self):
        return f"Series (title={self.title} year={self.year}, genre={self.genre}, episode={self.episode}, season={self.season}, views={self.views})"
  
             

# Niech program po uruchomieniu działa w następujący sposób:

# Wyświetli na konsoli komunikat Biblioteka filmów.
# Wypełni bibliotekę treścią.
# Wygeneruje odtworzenia treści za pomocą funkcji generate_views.
# Wyświetli na konsoli komunikat Najpopularniejsze filmy i seriale dnia <data>, gdzie <data> to bieżąca data w formacie DD.MM.RRRR.
# Wyświetli listę top 3 najpopularniejszych tytułów.            
              
if __name__ =="__main__":
    print("Biblioteka Filmów")
    movie0 = Movie(title="Sorority Babes in the Slimeball Bowl-O-Rama", year=1991, genre="comedy", views=3)
    movie1 = Movie(title="Tammy and the T-Rex", year=2019, genre="western", views=0)
    movie2 = Movie(title="Peanut Butter or Murder?", year=1976, genre="thiller", views=3)
    movie3 = Movie(title="Basketball Cats", year=1999, genre="drama", views=0)
    movie4 = Movie(title="Good Will Hunting", year=2005, genre="comedy", views=3)
    movie5 = Movie(title="Cannibal Women in the Avocado Jungle of Death", year=1998, genre="horror", views=0)
    movie6 = Movie(title="Paul Blart: Mall Cop", year=1994, genre="comedy", views=3)
    movie7 = Movie(title="Ghost Babies", year=2011, genre="western", views=0)
    movie8 = Movie(title="The Bye Bye Man", year=1971, genre="comedy", views=3)
    movie9 = Movie(title="Vampire Friends", year=2009, genre="western", views=0)
    series0 = Series(title="Astronaut Problems", year=2018, genre="documentary", episode=4, season=3, views=9)
    series1 = Series(title="I Bought a Vampire Motorcycle", year=1987, genre="comedy", episode=4, season=3, views=9)
    series2 = Series(title="A Shoe Addict’s Christmas", year=1999, genre="drama", episode=1, season=1, views=9)
    series3 = Series(title="The Dog Wedding", year=2017, genre="thiller", episode=2, season=3, views=9)
    series4 = Series(title="The World’s Best Dog Adventure", year=2005, genre="horror", episode=5, season=9, views=9)
    series5 = Series(title="Death Bed: The Bed That Eats", year=1887, genre="comedy", episode=6, season=9, views=9)
    series6 = Series(title="Attack of the Killer Tomatoes", year=1987, genre="drama", episode=3, season=6, views=9)
    series7 = Series(title="Operation Dumbo Drop", year=1234, genre="thiller", episode=5, season=8, views=9)
    series8 = Series(title="Revenge of the Bees", year=6577, genre="thiller", episode=4, season=8, views=9)
    series9 = Series(title="Home Page", year=2020, genre="comedy", episode=4, season=7, views=9)
     
    Movie.generate_views()       
            
    today = date.today()
    today = today.strftime("%d.%m.%Y")
    top = Movie.top_titles(3)  
    print("Najpopularniejsze filmy i seriale dnia:", today, top)
```
