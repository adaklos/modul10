```python
#Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000.
# Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. 
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mu, sigma = 3, 1 # średnia i odchylenie standardowe
zbior_a = np.random.normal(mu, sigma, 100)
zbior_b = np.random.normal(mu, sigma, 10000)
```

```python
# mu=3, sigma=1, 100

mean_a = np.round(np.mean(zbior_a))
median_a = np.median(zbior_a)
q0 = np.quantile (zbior_a,0.0)
q1 = np.round(np.quantile(zbior_a,0.25))
q2 = median_a
q3 = np.round(np.quantile(zbior_a,0.75))
q4 = np.quantile (zbior_a,1.0)
print(f'średnia:{mean_a}')
print(f'mediana:{median_a}')
print(f'Q0:{q0}')
print(f'Q1:{q1}')
print(f'Q2:{q2}')
print(f'Q3:{q3}')
print(f'Q4:{q4}')
```

```python
# mu=3, sigma=1, 10000

mean_b = np.round(np.mean(zbior_b))
median_b = np.median(zbior_b)
q0b = np.quantile (zbior_b,0.0)
q1b = np.round(np.quantile(zbior_b,0.25))
q2b = median_b
q3b = np.round(np.quantile(zbior_b,0.75))
q4b = np.quantile (zbior_b,1.0)
print(f'średnia:{mean_b}')
print(f'mediana:{median_b}')
print(f'Q0:{q0b}')
print(f'Q1:{q1b}')
print(f'Q2:{q2b}')
print(f'Q3:{q3b}')
print(f'Q4:{q4b}')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(10, 4))

axes[0].hist(zbior_a, bins=25,alpha=0.7)
axes[0].axvline(x = mean_a, color='red', label = 'Średnia', lw=3.0)
axes[0].axvline(x = median_a, color='green', label = 'Mediana')
axes[0].axvline(x = q0, color='black', label = 'Min', )
axes[0].axvline(x = q1, color='black', label = 'Q1')
axes[0].axvline(x = q3, color='black', label = 'Q3')
axes[0].axvline(x = q4, color='black', label = 'Max')
axes[0].set_title('mu=3, sigma=1, 100')
axes[0].legend(loc = 0)


axes[1].hist(zbior_b, bins=25, alpha=0.7)
axes[1].axvline(x = mean_b, color='red', label = 'Średnia', lw=3.0)
axes[1].axvline(x = median_b, color='green', label = 'Mediana')
axes[1].axvline(x = q0b, color='black', label = 'Min', )
axes[1].axvline(x = q1b, color='black', label = 'Q1')
axes[1].axvline(x = q3b, color='black', label = 'Q3')
axes[1].axvline(x = q4b, color='black', label = 'Max')
axes[1].set_title('mu=3, sigma=1, 10000')
axes[1].legend(loc = 0)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(10, 4))

axes[0].boxplot(zbior_a)
axes[0].set_title('mu=3, sigma=1, 100')
axes[1].boxplot(zbior_b)
axes[1].set_title('mu=3, sigma=1, 10000')
```
